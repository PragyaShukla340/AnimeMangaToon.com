## Why is the Tower of God Show So Popular?

```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

### Hypothetical Data

```python
random.seed(42)
days=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
page_views=np.random.randint(10000,20000,7)
average_time_spent=np.random.randint(8000,15000,7)
bounce_rate=np.random.randint(20,50,7)

Tower_of_God=pd.DataFrame({"Days":days,"Page_Views":page_views,
                           "Average_Time_Spent":average_time_spent,
                           "Bounce_Rate":bounce_rate})


Tower_of_God
```

```python
x_values=Tower_of_God["Days"]
y_values=Tower_of_God["Page_Views"].sort_values(ascending=True)
plt.figure(figsize=(8,6))
plt.bar(x_values,y_values,color="purple")
plt.xlabel("Days of Week -------> ")
plt.ylabel("Page Views -------> ")
plt.title("Days v/s Page Views")
plt.show()
```

```python
label=Tower_of_God["Days"]
values=Tower_of_God["Bounce_Rate"]

plt.figure(figsize=(8,6))
plt.pie(values,labels=label,autopct="%1.2f%%",startangle=90)
plt.title("Day wise Bounce Rate")
plt.show()
```

```python
x_values=Tower_of_God["Days"]
y_values=Tower_of_God["Average_Time_Spent"]
plt.figure(figsize=(8,6))
plt.plot(x_values,y_values,marker="*")
plt.xlabel("Days -------> ")
plt.ylabel("Average_Time_Spent -------> ")
plt.title("Days v/s Average_Time_Spent")
# plt.grid(True)
plt.show()
```

### Correlation

```python
df=pd.DataFrame({"Page_Views":page_views,
                "Average_Time_Spent":average_time_spent,
                "Bounce_Rate":bounce_rate})

correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", square=True, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

### Hypothetical Data

```python
age_groups=["11-18","18-25","25-32","32+"]
user_types=["New_Visitors","Returning_Visitors"]
Demographic_Data=[]
Behaviour_Data=[]
for age_group in age_groups:
    visitors = random.randint(10000, 50000)  
    bounce_rate = round(random.uniform(35, 65), 2)  
    avg_session_duration = round(random.uniform(1.5, 3.5), 2) 
    returning_visitors = round(random.uniform(20, 60), 2) 
    comments = random.randint(100, 500)  
    shares = random.randint(100, 700) 
    
    Demographic_Data.append([age_group, visitors, bounce_rate, avg_session_duration, returning_visitors, comments, shares])

for user_type in user_types:
    visitors = random.randint(10000, 50000)  
    bounce_rate = round(random.uniform(20, 70), 2)  
    avg_session_duration = round(random.uniform(1.5, 4.0), 2)  
    comments = random.randint(100, 700)  
    shares = random.randint(100, 1100)  
    
    Behaviour_Data.append([user_type, visitors, bounce_rate, avg_session_duration, comments, shares])

demographic_df = pd.DataFrame(Demographic_Data, columns=['Age Group', 'Visitors', 'Bounce Rate (%)', 'Avg Session Duration (min)', 'Returning Visitors (%)', 'Comments', 'Shares'])
behaviour_df = pd.DataFrame(Behaviour_Data, columns=['Visitor Type', 'Visitors', 'Bounce Rate (%)', 'Avg Session Duration (min)', 'Comments', 'Shares'])

demographic_df    
```

```python
behaviour_df
```

### Demographic Data Visualization

```python
label=demographic_df["Age Group"]
values=demographic_df["Returning Visitors (%)"]

plt.figure(figsize=(8,6))
plt.pie(values,labels=label,autopct="%1.2f%%",startangle=90)
plt.title("Returning Visitors(%) for different age groups")
plt.show()
```

```python
x_values = demographic_df["Age Group"]
y_values = demographic_df["Avg Session Duration (min)"]

plt.bar(x_values,y_values)

plt.xlabel("Age Group -------> ")
plt.ylabel("Avg Session Duration (min) -------> ")
plt.title("time spent in (min) by diffrent age groups ")
plt.figure(figsize=(5,5))

plt.show()
```

```python
x_values = demographic_df["Age Group"]
y_values = demographic_df["Bounce Rate (%)"]

plt.plot(x_values,y_values,marker="o")

plt.xlabel("Age Group -------> ")
plt.ylabel("Bounce Rate (%) -------> ")
plt.title(" Age Group to Bounce Rate(%) ")
plt.figure(figsize=(5,6))

plt.show()
```

```python
x_values = demographic_df["Age Group"]
y_values = demographic_df['Comments']
y_1_values = demographic_df["Shares"]

bar_width = 0.3
r1 = np.arange(len(x_values))  
r2 = [x + bar_width for x in r1] 

plt.figure(figsize=(10,6))
bars_1 = plt.bar(r1, y_values, width=bar_width, label="Comments")
bars_2 = plt.bar(r2, y_1_values, width=bar_width, label="Shares")
plt.xlabel("Age Groups -------> ")
plt.ylabel("Values -------> ")
plt.title("Visitors and Their Comments & shares")
plt.xticks([r + bar_width/2 for r in range(len(x_values))], x_values)
plt.legend()
plt.show()
```
